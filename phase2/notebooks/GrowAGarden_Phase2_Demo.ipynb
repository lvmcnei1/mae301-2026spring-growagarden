# Grow A Garden - Phase 2 Colab Demo

This notebook demonstrates the dataset-driven planning engine built for Phase 2. It generates crop recommendations and an 8-week planting/watering schedule.

```
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT / 'phase2'
sys.path.insert(0, str(PROJECT_ROOT / 'src'))
PROJECT_ROOT
```

```
!pip -q install -r {PROJECT_ROOT / 'requirements.txt'}
```

```
from growagarden.data_loader import build_plant_profiles
from growagarden.scheduler import plan_garden
from growagarden.weather import load_weather_csv

build_plant_profiles(save=True)
weather_df = load_weather_csv(PROJECT_ROOT / 'data' / 'sample_weather' / 'phoenix_week.csv')
result = plan_garden(
    planning_date='2026-04-12',
    zone='9b',
    garden_area_sqft=48,
    sun_hours=8,
    watering_preference='medium',
    selected_crops=['Tomatoes', 'Basil', 'Beans', 'Bell Peppers'],
    weather_df=weather_df,
)
result['recommendations'][['name','score','action_now','water_inches_per_week','estimated_harvest_date']]
```
